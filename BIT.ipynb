```python
#bitwise operators


a = 5 # In binary: 0101
b = 3 # In binary: 0011
print(a & b)  # Bitwise AND
print(a | b)  # Bitwise OR
print(a ^ b)  # Bitwise XOR
print(~a)     # Bitwise NOT
```

```python
#left shift
a = 5 # In binary: 0101
print(a << 1)  # Left shift by 1 (0101 becomes 1010, which is 10 in decimal)
#right shift
a = 5 # In binary: 0101
print(a >> 1)  # Right shift by 1 (0101 becomes 0010, which is 2 in decimal)
```

```python

num1 = float(input("Enter first number: "))
num2 = float(input("Enter second number: "))

print("Choose operation: add, subtract, multiply, divide")
operation = input("Enter operation: ")

if operation == "add":
    print("Result:", num1 + num2)
elif operation == "subtract":
    print("Result:", num1 - num2)
elif operation == "multiply":
    print("Result:", num1 * num2)
elif operation == "divide":
    if num2 != 0:
        print("Result:", num1 / num2)
    else:
        print("Error: Division by zero not allowed")
else:
    print("Invalid operation")
```

```python
# Python String Manipulation Examples

# Define the original string
s = "Hello, world!"

# Convert the string to uppercase
print(s.upper())    # Output: 'HELLO, WORLD!'

# Convert the string to lowercase
print(s.lower())    # Output: 'hello, world!'

# Remove leading and trailing whitespaces from the string
print(s.strip())    # Output: 'Hello, world!'

# Replace all occurrences of 'o' with 'x' in the string
print(s.replace('o', 'x'))    # Output: 'Hellx, wxrld!'

# Count the number of occurrences of 'a' in the string
print('Abracadabra'.count('a'))    # Output: 4
```

```python
name = "Alice"
age = 30

# 1. Using the '%' operator (old-style formatting)
print("My name is %s and I am %d years old." % (name, age))

# 2. Using the 'format()' method
print("My name is {} and I am {} years old.".format(name, age))

# 3. Using f-strings (Python 3.6+)
print(f"My name is {name} and I am {age} years old.")
```

```python
# Newline
print("Hello\nWorld")  
# Output:
# Hello
# World

# Tab
print("Hello\tWorld")  
# Output: Hello    World

# Single quote inside single quotes
print('It\'s Python')  
# Output: It's Python

# Double quote inside double quotes
print("He said \"Python is fun\"")  
# Output: He said "Python is fun"

# Backslash itself
print("C:\\Users\\Navaneeth")  
# Output: C:\Users\Navaneeth

# Bell sound (may not work everywhere)
print("\a")  

# Unicode character
print("\u03A9")   # Output: Ω (Greek Omega)
```
