```python
#implicit type conversion
num_int = 5       # int
num_float = 2.5   # float

result = num_int + num_float
print(result)        # 7.5
print(type(result))  # <class 'float'>
```

```python
#explicit type conversion
# String → Integer
num_str = "100"
num_int = int(num_str)
print(num_int + 50)   # 150

# Float → Integer
num_float = 9.8
num_int = int(num_float)
print(num_int)        # 9 (decimal part removed)

# Integer → String
num = 42
num_str = str(num)
print("Answer is " + num_str)  # Answer is 42

# String → Float
num_str = "3.14"
num_float = float(num_str)
print(num_float + 1)  # 4.14
```
